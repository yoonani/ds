본 문서는 from An Introduction to R for Spatial Analysis & Mapping(2015, Chris Brudson, Lex Comber, Sage)의 43페이지에서 44페이지 사이의 코드를 적용한 것입니다.

### 사용하는 패키지
`GISTools`, `RColorBrewer`를 사용하므로 다음과 같이 설치하고 작업공간으로 불러옵니다. (이미 설치되어 있다면 다음의 `install.packages()`는 입력하지 않으셔도 됩니다.)

```r
install.packages(c("GISTools", "RColorBrewer"), dependencies=TRUE)
```

```r
library(GISTools)
library(RColorBrewer)
```

`sp` 패키지에서 예제 데이터로 제공하는 meuse.grid 데이터를 사용합니다.

```r
data(meuse.grid)
```

```r
str(meuse.grid)
```

```r
mat <- SpatialPixelsDataFrame(points=meuse.grid[c("x", "y")], data=meuse.grid)
```

```r
defpar <- par(no.readonly=TRUE)
par(mar=c(0, 0, 0, 0))
```

R의 기본 패키지인 `graphics`에 있는 `image()`함수는 색칠된 격자를 만들어 줍니다.

```r
image(mat, "dist")
```

`RColorBrewer`가 제공하는 팔레트를 사용합니다.
아래의 코드는 "Greens" 팔레트로 부터 7개의 색을 가져와 변수 `grPal`에 저장합니다.

```r
grPal <- brewer.pal(7, "Greens")
```

위에서 만든 팔레트를 적용합니다.

```r
image(mat, "dist", col=grPal)
```

앞서 그래프의 여백을 0으로 만든 설정을 원래대로 돌리기 위해 기본 설정값을 저장한 변수를 이용합니다.

```r
par(defpar)
```
